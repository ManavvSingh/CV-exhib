```python
import cv2
import numpy as np
from pyzbar.pyzbar import decode

# Capture video from the default camera
cap = cv2.VideoCapture(0)
cap.set(3, 640)  # Set width
cap.set(4, 480)  # Set height

# Read authorized data from a file
with open('myDataFile.text') as f:
    myDataList = f.read().splitlines()

# Main loop to process video frames
while True:
    # Read a frame from the video capture
    success, img = cap.read()
    
    # Decode QR codes in the frame
    for barcode in decode(img):
        # Get decoded data from the barcode
        myData = barcode.data.decode('utf-8')
        

        # Check if the decoded data is in the authorized list
        if myData in myDataList:
            myOutput = 'Authorized'
            myColor = (0, 255, 0)  # Green color for authorized
        else:
            myOutput = 'Un-Authorized'
            myColor = (0, 0, 255)  # Red color for unauthorized

        # Draw polygon around the barcode
        pts = np.array([barcode.polygon], np.int32)
        pts = pts.reshape((-1, 1, 2))
        cv2.polylines(img, [pts], True, myColor, 5)

        # Draw text for authorization status and decoded data
        pts2 = barcode.rect
        cv2.putText(img, myOutput, (pts2[0], pts2[1]), cv2.FONT_HERSHEY_SIMPLEX,
                    0.9, myColor, 2)
        cv2.putText(img, myData, (pts2[0], pts2[1] - 30), cv2.FONT_HERSHEY_SIMPLEX,
                    0.5, myColor, 1)  # Display decoded data above barcode

    # Display the result image
    cv2.imshow('Result', img)
    
    # Wait for a key press, exit if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close all windows
cap.release()
cv2.destroyAllWindows()

```
